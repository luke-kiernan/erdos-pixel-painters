```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# from the CIFAR10 dataset readme.
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
# goal: save the whole training dataset as one tensor, so that batches just return views to said tensor
# load and convert to grayscale.
batch1 = unpickle('./data/cifar-10-batches-py/data_batch_1')[b'data'].reshape(-1,3,32,32)
batch1Grayscale = np.dot(np.transpose(batch1, (0,2,3,1)), [0.2989, 0.5870, 0.1140])
```

```python
f, ax = plt.subplots(2)
ax[0].imshow(np.transpose(batch1[1], (1, 2, 0)))
ax[1].imshow(batch1Grayscale[1], cmap='gray')
for a in ax:
    a.axis("off")

# next step would be to turn the uint8's into normalized floats, probably via numpy array manipulations.
# But I suspect my original goal no longer makes sense:
# (1) Examples online store lists. Surely if using one big tensor/numpy array was faster, people would do that.
# (2) See https://stackoverflow.com/questions/52818145/why-pytorch-dataloader-behaves-differently-on-numpy-array-and-list
```
